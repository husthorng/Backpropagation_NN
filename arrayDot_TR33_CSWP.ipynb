```python
worksheet = gc.open("2025_DT hung").worksheet("tr33W1")
rows = worksheet.get_all_values()
print(rows)
records = pd.DataFrame(rows) #轉pandas結構

rows.pop(0)    #刪除第一列
W1 = np.array(rows)
W1 = W1.astype(float)

W1
```

```python
L0s=np.dot(test, W0.T)
L0s
```

```python
def nonlin(x,deriv=False):
    A=1/(1+np.exp(-x))
    if(deriv==True):
        return (A*(1-A))
    return A
```

```python
L0=nonlin(L0s)
L0
```

```python
L1s=np.dot(L0, W1.T)
L1s
```

```python
L1=nonlin(L1s)
L1
```

```python
result = (L1 > 0.9).astype(int)
result
```
